## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Проанализируем стоимость жилья в зависимости от количества жилой площади в домах. Построим график:

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg') 
```

Вывод: на данной диаграмме мы видим, что количество квадратных метров в помещении напрямую влияет на стоимость недвижимости: чем боле площадь, тем выше стоимость

Далее посмотрим как влияет оценка состояния дома на его стоимость

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['condition'], y=df['price'], whis=1.5, showfliers=False)

plt.xlabel('condition')
plt.ylabel('price')
plt.title('Р');
```

На боксплоте видно, что чем лучше состояние дома, тем выше его средняя стоимость, хотя дома, у которых оценка хорошо и удовлетворительно имеют примерно одинаковую стоимость. 

Создадим матрицу корреляций, которая наглядно покажет зависимость одних показателей от других:

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Благодаря даннйо матрице мы можем легко увидеть, что на стоимость недвижимости сильнее всего влияют жилая площадь, уровень конструкции здания, количество ванных комнат, жилая площадь над уровнем земли, а такие показатели как этажность, набережная, вид, количество спален оказывают меньшее влияние на стоимость жилья.
